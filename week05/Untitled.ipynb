```python
import math
import matplotlib.pyplot as plt
from numpy import arange, linspace
%matplotlib inline

def Ek(kx,ky,kz,V,Es):
    return Es+4.0*V*(math.cos(kx/2.0)*math.cos(ky/2.0)+math.cos(ky/2.0)*math.cos(kz/2.0)+math.cos(kz/2.0)*math.cos(kx/2.0))

sample=100
Vss=-0.2
Ess=-1.0
GammaPoint=[0.0,0.0,0.0]
XPoint=[2.0*math.pi,0.0,0.0]
kpath=[]
kpath.append(GammaPoint)
kpath.append(XPoint)
kx=[]
energy=[]
for i in range(len(kpath)-1):
    for j in range(sample):
        kpointx=kpath[i][0]+(kpath[i+1][0]-kpath[i][0])/float(sample)*float(j)
        kpointy=kpath[i][1]+(kpath[i+1][1]-kpath[i][1])/float(sample)*float(j)
        kpointz=kpath[i][2]+(kpath[i+1][2]-kpath[i][2])/float(sample)*float(j)
        kx.append(kpointx)
        energy.append(Ek(kpointx,kpointy,kpointz,Vss,Ess))

plt.plot(kx,energy)
plt.xlabel("kx")
plt.ylabel("energy")
plt.show()
```

```python
import math 
import matplotlib.pyplot as plt
from numpy import arange,linspace
%matplotlib inline

def gaussian(width, var):
    prefactor=1.0/width/math.sqrt(2.0*math.pi)
    return prefactor*math.exp(-var*var/2.0/width/width)

def Ek(kx, ky, V, Es):
    return Es+2.0*V*(math.cos(kx))+2.0*V*(math.cos(ky))

sample=400
Vss=-0.1
Ess=0
Esample=100
GaussianWidth=0.01
kmesh=[]
kx=linspace(-math.pi,math.pi,sample)
ky=linspace(-math.pi,math.pi,sample)


energy=[]
for i in kx:
    for j in ky:
        energy.append(Ek(i,j,Vss,Ess))

Eval=linspace(-1,1,Esample)
dos=[]
for i in Eval:
    value=0.0
    for j in energy:
        value=value+gaussian(GaussianWidth,j-i)/float(len(energy))
    dos.append(value)
    
plt.plot(Eval,dos)
plt.show()
```
